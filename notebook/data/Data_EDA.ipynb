```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.datasets import load_diabetes
```

```python
diabetes= load_diabetes()
```

```python
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df.head()
```

```python
df.info()
```

```python
print("\n=== FEATURE DESCRIPTIONS ===")
feature_descriptions = {
    'age': 'Age in years',
    'sex': 'Gender (1 = male, -1 = female)', 
    'bmi': 'Body mass index',
    'bp': 'Average blood pressure',
    's1': 'Total cholesterol (tc)',
    's2': 'Low-density lipoproteins (ldl)', 
    's3': 'High-density lipoproteins (hdl)',
    's4': 'Total cholesterol / HDL ratio (tch)',
    's5': 'Logarithm of serum triglycerides (ltg)',
    's6': 'Blood sugar level (glu)'
}
for feature, description in feature_descriptions.items():
    print(f"{feature}: {description}")
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(15,10))
```

```python
df['target']=diabetes.target
```

```python
print("=== DIABETES DATASET OVERVIEW ===")
print(f"Dataset shape: {df.shape}")
print(f"Features: {diabetes.feature_names}")
print(f"Target: Disease progression (higher = more progression)")
```

```python
print("\n=== TARGET DISTRIBUTION ===")
print(f"Target min: {df['target'].min()}")
print(f"Target max: {df['target'].max()}")
print(f"Target mean: {df['target'].mean():.2f}")
print(f"Target std: {df['target'].std():.2f}")
```

```python
plt.figure(figsize=(8, 10))
plt.subplot(2, 3, 1)
plt.hist(df['target'], bins=30, alpha=0.7, color='skyblue')
plt.title('Distribution of Disease Progression')
plt.xlabel('Disease Progression Score')
plt.ylabel('Frequency')
```

```python
# 2. Correlation heatmap
plt.subplot(1, 2, 2)
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
```

```python
target_corr= df.corr()['target'].abs().sort_values(ascending=False)
top_features= target_corr.head(5).index[1:]
```

```python
for i, feature in enumerate(top_features):
    plt.subplot(2,3,i+3)
    plt.scatter(df[feature],df['target'],alpha=0.6)
    plt.xlabel(feature)
    plt.ylabel('Disease Progression')
    plt.title(f"{feature} vs Target (corr: {df[feature].corr(df["target"]):.3f})")

plt.tight_layout()
plt.show()    
```

```python
print("\n=== KEY INSIGHTS ===")
print("1. All features are already normalized (mean ~0, std ~1)")
print("2. No missing values - clean dataset!")
print("3. Target ranges from 25 to 346 (disease progression score)")
print("4. Most correlated features with target:")
for feature in target_corr.head(5).index[1:]:
    print(f"   - {feature}: {target_corr[feature]:.3f}")
    
```
